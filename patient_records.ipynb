```python
# saving Patient records by using pandas
# patent registaration by using pandas
import pandas as pd
from csv import writer
```

```python
patient = {'patient_id':[],
'Name':[],'age':[],'gender':[],'issue_related':[],'Doctor_suggested':[]}

df = pd.DataFrame(patient)
df.to_csv("record1.csv")
```

```python
i = int(input("Enter index:"))
pi = int(input("Enter Patient ID:"))
n = input("Enter Patient Name:")
a = int(input("Enter Age Of Patient:"))
g = input("Enter Gender:")
h = input("Issue Related To:")
d = input("Enter Name of Docter Sugessted")
p_records = [i,pi,n,a,g,h,d]
with open('record1.csv', 'a') as f_object:
  
    # Pass this file object to csv.writer()
    # and get a writer object
    writer_object = writer(f_object)
  
    # Pass the list as an argument into
    # the writerow()
    writer_object.writerow(p_records)
  
    #Close the file object
    f_object.close()
```
