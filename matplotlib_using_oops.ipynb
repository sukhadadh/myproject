```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import pandas as pd
```

```python
df = pd.read_csv("/content/Height of Male and Female by Country 2022.csv")
```

```python
df.columns
```

```python
df.shape
```

```python
nt = df.head(5)
nt
```

```python
class graphs():
  def __init__(self,nt):
    self.nt= nt
  def LinePlot(self):
    plt.figure(figsize = (16,8))
    plt.plot(self.nt['Country Name'],self.nt['Male Height in Cm'])
    plt.title("Heght of males by country")
    plt.xlabel("Country Name")
    plt.ylabel("Male height in cm")
    plt.show()
    return None

  def BarPlot(self):
    plt.figure(figsize =(16,8))
    plt.bar(self.nt['Male Height in Ft'],self.nt['Female Height in Ft'],width = 0.1)
    plt.title("illustration of bar")
    plt.xlabel("male height in ft")
    plt.ylabel("female height in ft")
    plt.show()
    return None
  
  def HistPlot(self):
    plt.figure(figsize =(16,8))
    plt.hist(self.nt['Female Height in Cm'])
    plt.title("histogram ittustration")
    plt.show()

  def Scatterplot(self):
    plt.figure(figsize=(16,8))
    plt.scatter(self.nt['Male Height in Ft'],self.nt['Female Height in Ft'], c = 'green')
    plt.title("scatter plot example")
    plt.show()
  
  def PiePlot(self):
    plt.figure(figsize=(16,8))
    plt.pie(self.nt['Female Height in Ft'])
    plt.title(self.nt['Country Name'][0:5])
    plt.show()

  def image(self):
    testImage = img.imread('sample.jpg')
    plt.imshow(testImage)
```

```python
g = graphs(nt =nt)
g.LinePlot()
```

```python
g = graphs(nt =nt)
g.BarPlot()
```

```python
g.HistPlot()
```

```python
g.Scatterplot()
```

```python
g.PiePlot()
```

```python
g.image()
```
